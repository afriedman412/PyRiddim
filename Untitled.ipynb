```python
dicto = {
    'artist':[],
    'tune':[],
    'riddim':[],
    'label':[],
    'year':[],
    'producer':[],
    'album':[]
}
```

```python
cats = ['artist','tune','riddim','label','year','producer','album']
```

```python
for row in soup.find_all('tr', {'class':'info-cell'}):
    for cat, cell in zip(cats, row.find_all('td')):
        if cell.find('a') is not None:
            d = cell.find('a').text
        else:
            d = cell.text
        dicto[cat].append(d)
```

```python
for row in soup.find_all('td', {'class':'info-cell'})[0:14]:
    print(row)
    for cat, cell in zip(cats, row.find_all('td')):
        if cell.find('a') is not None:
            d = cell.find('a').text
        else:
            d = cell.text
        dicto[cat].append(d)
```

```python
for k in dicto.keys():
    print(len(dicto[k]))
```
